```python
#For Utilities, all the values are pretty much the same, so it won't affect our model much
train = train.drop(['Utilities'], axis=1)
```

```python
#Check for missing data (how many missing observations, out of 1460)
pd.concat([train.isnull().sum().sort_values(ascending=False),(train.isnull().sum()*100/train.isnull().count()).sort_values(ascending=False)],axis=1,keys=['Total','Percentage'])
```

It seems that four of our columns are missing over 80% of their data. However, after looking at the variable descriptions, it turns out that most of these aren't actually missing data. If the house doesn't have a pool, then Pool Quality will be left empty. SO for many of the 'missing data' we need to fill them with 'none.'

```python
#Filling in 'None' where we're missing categorial data, because the feature doesn't exist at that specific property
for col in ('MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', "MasVnrType", 'MSSubClass'):
    train[col]=train[col].fillna('None')
```

```python
#Similarly, there are variables where missing data means 0, so let's fill those
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea', 'GarageCars',
               'GarageArea', 'GarageYrBlt'):
    train[col]=train[col].fillna(0)
```

```python
#We can use the median LotFrontage for each respective neighborhood to fill in missing values for LotFrontage
train["LotFrontage"] = train.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
#Let's check and see what missing data is left
pd.concat([train.isnull().sum().sort_values(ascending=False),(train.isnull().sum()*100/train.isnull().count()).sort_values(ascending=False)],axis=1,keys=['Total','Percentage'])
```

```python
#Let's look at Electrical
train['Electrical'].value_counts()
```

```python
#The large majority of houses have Standard Circuit Breakers, so let's just fill the missing row with that
train['Electrical'] = train['Electrical'].fillna('SBrkr')
```
