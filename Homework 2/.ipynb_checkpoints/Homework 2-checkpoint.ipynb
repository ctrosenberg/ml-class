### Missing Data

```python
#Check for missing data (how many missing observations, out of 1460)
pd.concat([train.isnull().sum().sort_values(ascending=False),(train.isnull().sum()*100/train.isnull().count()).sort_values(ascending=False)],axis=1,keys=['Total','Percentage'])
```

It seems that four of our columns are missing over 80% of their data. However, after looking at the variable descriptions, it turns out that most of these aren't actually missing data. If the house doesn't have a pool, then Pool Quality will be left empty. SO for many of the 'missing data' we need to fill them with 'none.'

```python
#Filling in 'None' where we're missing categorial data, because the feature doesn't exist at that specific property
for col in ( 'PoolQC','MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', "MasVnrType", 'MSSubClass'):
    train[col]=train[col].fillna('None')
    test[col]=test[col].fillna('None')
print(train.shape,test.shape)
```

```python
#Similarly, there are variables where missing data means 0, so let's fill those
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea', 'GarageCars',
               'GarageArea', 'GarageYrBlt'):
    train[col]=train[col].fillna(0)
    test[col]=test[col].fillna(0)
    
print(train.shape,test.shape)
```

```python
#We can use the median LotFrontage for each respective neighborhood to fill in missing values for LotFrontage
train["LotFrontage"] = train.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
test["LotFrontage"] = test.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
print(train.shape,test.shape)
```

```python
#Let's check and see what missing data is left
pd.concat([train.isnull().sum().sort_values(ascending=False),(train.isnull().sum()*100/train.isnull().count()).sort_values(ascending=False)],axis=1,keys=['Total','Percentage'])
```

```python
#Let's look at Electrical
train['Electrical'].value_counts()
print(train.shape,test.shape)
```

```python
#The large majority of houses have Standard Circuit Breakers, so let's just fill the missing row with that
train['Electrical'] = train['Electrical'].fillna('SBrkr')
test['Electrical'] = test['Electrical'].fillna('SBrkr')
print(train.shape,test.shape)
```

```python
#Last call for missing data
train.isnull().any().sum()
#print(train.shape,test.shape)
```

Woohoo, no more missing data on the training set!

### Categorical to Numerical

```python
#We need to transform the variables with hierarchal categorical data into numerical data 
train['MSSubClass'] = train['MSSubClass'].apply(str)
train['OverallCond'] = train['OverallCond'].astype(str)
train['YrSold'] = train['YrSold'].astype(str)
train['MoSold'] = train['MoSold'].astype(str)
#Test set
test['MSSubClass'] = test['MSSubClass'].apply(str)
test['OverallCond'] = test['OverallCond'].astype(str)
test['YrSold'] = test['YrSold'].astype(str)
test['MoSold'] = test['MoSold'].astype(str)
print(train.shape,test.shape)
```

```python
#Let's now create dummy variables for the other categorical variables
train = pd.get_dummies(train)
test = pd.get_dummies(test)
print(train.shape,test.shape)
```

```python
#Let's create our target variable and drop SalePrice
target = train['SalePrice']
train.drop(['SalePrice'], axis=1, inplace=True)

#Now we need to make sure our sets are the same shape
test.drop(['MSSubClass_150'], axis=1, inplace=True)
test_columns = test.columns
train = train[test_columns]
print(train.shape,test.shape)
```

## Linear Regression Model

```python
#Let's scale our data
scaler = StandardScaler()
scaler.fit(train)
scaled = scaler.transform(train)
```

```python
#Let's split the training data into a training and test
X_train, X_test, y_train, y_test = train_test_split(scaled, target, test_size=.2)
#Let's run the Linear Regression
linear_regression_model = SGDRegressor(tol=.0001, eta0=.01)
linear_regression_model.fit(X_train, y_train)
train_predictions = linear_regression_model.predict(X_train)
test_predictions = linear_regression_model.predict(X_test)

train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

print("Train RMSE: {}".format(np.sqrt(train_mse)))
print("Test RMSE: {}".format(np.sqrt(test_mse)))
```

Wow, the linear regression isn't looking too hot.

```python
#Let's try running K-nearest neighbors
reg = KNeighborsRegressor()
#gridsearch = GridSearchCV(reg, {"n_neighbors": [1, 3, 5, 7, 9, 11], "weights": ['uniform', 'distance'], 
#                                'p': [1, 2, 3]}, scoring='neg_mean_squared_error')
gridsearch = GridSearchCV(reg, {"n_neighbors": [1, 3, 5, 7, 9, 11]})
gridsearch.fit(X_train, y_train)
train_predictions = gridsearch.predict(X_train)
test_predictions = gridsearch.predict(X_test)

train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

print("Best Params: {}".format(gridsearch.best_params_))
print("Train RMSE: {}".format(np.sqrt(train_mse)))
print("Test RMSE: {}".format(np.sqrt(test_mse)))
```

```python
KNN = KNeighborsRegressor(n_neighbors=7)
KNN.fit(train,target)
predictions = KNN.predict(test)

#save results to csv to submit to kaggle
knn_outcome = pd.DataFrame({"ID":test_Id,"SalePrice":np.expm1(predictions)})
knn_outcome.to_csv("KNN.csv",index=False)
knn_outcome.head()
```

Alright! The K-Nearest Neighbors model scored 0.23705 on Kaggle, a little better than our 1 variable model. 

```python
#Let's try Lasso
lassy = Lasso(alpha=0.0005, max_iter=100, selection='random')

lassy.fit(X_train, y_train)
train_predictions = lassy.predict(X_train)
test_predictions = lassy.predict(X_test)

train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

print("Train RMSE: {}".format(np.sqrt(train_mse)))
print("Test RMSE: {}".format(np.sqrt(test_mse)))
```

Our lasso model did better than our K-Nearest Neighbors, but let's try another model.

```python
#Let's try running an ElasticNetCV model
clf = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1], alphas=[.001,.01,.1, 1, 10, 100])
clf.fit(X_train, y_train)
train_predictions = clf.predict(X_train)
test_predictions = clf.predict(X_test)

train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

print("Train RMSE: {}".format(np.sqrt(train_mse)))
print("Test RMSE: {}".format(np.sqrt(test_mse)))
```

```python
clf.fit(train,target)
predictions = clf.predict(test)

#save results to csv to submit to kaggle
knn_outcome = pd.DataFrame({"ID":test_Id,"SalePrice":np.expm1(predictions)})
knn_outcome.to_csv("ElasticNetCV.csv",index=False)
knn_outcome.head()
```
